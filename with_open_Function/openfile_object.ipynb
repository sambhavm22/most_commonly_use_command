### Creating a new csv file using writer function

```python
import csv

with open("with_open_Function/netflix_titles.csv", 'r') as read_csv_file:
    reader_csv = csv.reader(read_csv_file)

    with open("with_open_Function/copy_netflix_titles.csv", 'w') as write_csv_file:
        writer_csv = csv.writer(write_csv_file, delimiter="\t")

        for line in reader_csv:
            writer_csv.writerow(line)
```
